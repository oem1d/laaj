```python
import os
import json
import time
import numpy as np
import pandas as pd

from tqdm import tqdm
from datasets import load_dataset
from openai import OpenAI

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier


client = OpenAI()


AXIS = "helpfulness"      # helpfulness, correctness, complexity 중 선택
ALPHA = 0.1
N_SOURCE = 300           # 처음엔 작게. 비용 줄이기용
N_TARGET = 300
RANDOM_STATE = 0


Z_QUESTIONS = [
    "Is the response about the requested task? Choose one: wrong, related, requested.",
    "Does it provide the requested output type? Choose one: missing, partial, present.",
    "What simple task shape matters for helpfulness? Choose one: ordinary_or_interaction, structured_or_artifact, formal_reference_or_external.",
    "How many main parts are addressed? Choose one: many_missing, some_missing, covered.",
    "Are there concrete task-relevant details? Choose one: generic, some, concrete.",
    "Is there usable advice, instruction, or directly usable output? Choose one: none, partial, usable.",
    "Are important constraints followed? Choose one: violated, partially_followed, followed.",
    "Is there major hallucination or unsupported content? Choose one: major, minor, none.",
    "Is the response clear and organized? Choose one: unclear, mixed, clear.",
    "Overall, what is the internal judgment state? Choose one: poor, partial, good.",
]


def normalize_score(y):
    """
    HelpSteer 점수는 보통 0~4 또는 1~5 형태일 수 있으므로
    실제 min/max를 이용해 [0,1]로 변환.
    """
    return y.astype(float)


def build_text(row):
    prompt = row.get("prompt", "")
    response = row.get("response", "")
    return f"Prompt:\n{prompt}\n\nResponse:\n{response}"


def extract_z_with_llm(text, axis=AXIS, max_retries=3):
    questions = "\n".join([f"Z{i+1}. {q}" for i, q in enumerate(Z_QUESTIONS)])

    prompt = f"""
You are extracting low-cardinality guideline states for LLM-as-a-judge calibration.

Evaluation axis: {axis}

Input:
{text}

Answer the following questions.
Return ONLY valid JSON with keys Z1 to Z10.
Do not include explanation.

Questions:
{questions}
"""

    for _ in range(max_retries):
        try:
            res = client.chat.completions.create(
                model="gpt-4o-mini",
                messages=[{"role": "user", "content": prompt}],
                temperature=0,
            )
            content = res.choices[0].message.content.strip()
            return json.loads(content)
        except Exception:
            time.sleep(2)

    return {f"Z{i}": "unknown" for i in range(1, 11)}


def add_z_states(df, cache_path, axis=AXIS):
    if os.path.exists(cache_path):
        return pd.read_csv(cache_path)

    rows = []
    for _, row in tqdm(df.iterrows(), total=len(df)):
        text = build_text(row)
        z = extract_z_with_llm(text, axis=axis)

        new_row = row.to_dict()
        for i in range(1, 11):
            new_row[f"Z{i}"] = z.get(f"Z{i}", "unknown")

        rows.append(new_row)

    out = pd.DataFrame(rows)
    out.to_csv(cache_path, index=False)
    return out


class RoWCI:
    def __init__(self, z_cols, y_col, alpha=0.1, random_state=0):
        self.z_cols = z_cols
        self.y_col = y_col
        self.alpha = alpha
        self.random_state = random_state

    def fit(self, source_df, target_cov_df):
        d_src, d_cal = train_test_split(
            source_df,
            test_size=0.5,
            random_state=self.random_state,
            shuffle=True,
        )

        self.d_src = d_src.reset_index(drop=True)
        self.d_cal = d_cal.reset_index(drop=True)

        y_min = source_df[self.y_col].min()
        y_max = source_df[self.y_col].max()
        self.y_min = y_min
        self.y_max = y_max

        self.score_encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

        x_src = self.score_encoder.fit_transform(self.d_src[self.z_cols])
        y_src = self.d_src[self.y_col].astype(int).to_numpy()

        self.score_model = XGBClassifier(
            objective="multi:softprob",
            eval_metric="mlogloss",
            random_state=self.random_state,
        )
        self.score_model.fit(x_src, y_src)

        self.classes_ = self.score_model.classes_.astype(float)

        cal_score = self._state_score(self.d_cal[self.z_cols])
        cal_y = self._normalize_y(self.d_cal[self.y_col].to_numpy(dtype=float))
        self.cal_errors = np.abs(cal_score - cal_y)

        d0 = self.d_src[self.z_cols].copy()
        d0["A"] = 0

        d1 = target_cov_df[self.z_cols].copy()
        d1["A"] = 1

        d_dre = pd.concat([d0, d1], ignore_index=True)

        self.m0 = len(d0)
        self.m1 = len(d1)

        self.dre_encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
        x_dre = self.dre_encoder.fit_transform(d_dre[self.z_cols])
        y_dre = d_dre["A"].to_numpy()

        self.domain_model = RandomForestClassifier(
            random_state=self.random_state,
        )
        self.domain_model.fit(x_dre, y_dre)

        self.cal_weights = self._importance_weight(self.d_cal[self.z_cols])

        return self

    def predict_interval(self, test_df):
        s_test = self._state_score(test_df[self.z_cols])
        w_test = self._importance_weight(test_df[self.z_cols])

        cutoffs = np.array([
            self._weighted_conformal_cutoff(w)
            for w in w_test
        ])

        lower = np.clip(s_test - cutoffs, 0.0, 1.0)
        upper = np.clip(s_test + cutoffs, 0.0, 1.0)

        return pd.DataFrame({
            "sZ": s_test,
            "cutoff": cutoffs,
            "lower": lower,
            "upper": upper,
        })

    def evaluate(self, test_df):
        intervals = self.predict_interval(test_df)
        y_true = self._normalize_y(test_df[self.y_col].to_numpy(dtype=float))

        covered = (y_true >= intervals["lower"].to_numpy()) & (
            y_true <= intervals["upper"].to_numpy()
        )

        coverage = covered.mean()
        size = (intervals["upper"] - intervals["lower"]).mean()

        return coverage, size, intervals

    def _normalize_y(self, y):
        return (y - self.y_min) / (self.y_max - self.y_min)

    def _state_score(self, z_df):
        x = self.score_encoder.transform(z_df[self.z_cols])
        prob = self.score_model.predict_proba(x)

        class_values = (self.classes_ - self.y_min) / (self.y_max - self.y_min)
        return prob @ class_values

    def _importance_weight(self, z_df):
        x = self.dre_encoder.transform(z_df[self.z_cols])
        eta = self.domain_model.predict_proba(x)[:, 1]

        eta = np.clip(eta, 1e-6, 1 - 1e-6)
        return (self.m0 / self.m1) * eta / (1 - eta)

    def _weighted_conformal_cutoff(self, w_test):
        denom = np.sum(self.cal_weights) + w_test
        finite_weights = self.cal_weights / denom

        order = np.argsort(self.cal_errors)
        sorted_errors = self.cal_errors[order]
        sorted_weights = finite_weights[order]

        cdf = np.cumsum(sorted_weights)
        idx = np.searchsorted(cdf, 1 - self.alpha, side="left")

        if idx >= len(sorted_errors):
            return 1.0

        return sorted_errors[idx]


def main():
    hs1 = load_dataset("nvidia/HelpSteer")
    hs2 = load_dataset("nvidia/HelpSteer2")

    source = pd.DataFrame(hs1["train"]).sample(
        N_SOURCE, random_state=RANDOM_STATE
    ).reset_index(drop=True)

    target = pd.DataFrame(hs2["train"]).sample(
        N_TARGET, random_state=RANDOM_STATE
    ).reset_index(drop=True)

    needed_cols = ["prompt", "response", AXIS]
    source = source[needed_cols].rename(columns={AXIS: "Y"})
    target = target[needed_cols].rename(columns={AXIS: "Y"})

    source_z = add_z_states(source, f"hs1_z_{AXIS}_{N_SOURCE}.csv", axis=AXIS)
    target_z = add_z_states(target, f"hs2_z_{AXIS}_{N_TARGET}.csv", axis=AXIS)

    z_cols = [f"Z{i}" for i in range(1, 11)]

    model = RoWCI(
        z_cols=z_cols,
        y_col="Y",
        alpha=ALPHA,
        random_state=RANDOM_STATE,
    )

    model.fit(source_z, target_z)

    coverage, size, intervals = model.evaluate(target_z)

    print("==== RoWCI Result ====")
    print(f"Axis: {AXIS}")
    print(f"Coverage: {coverage:.3f}")
    print(f"Average interval size: {size:.3f}")
    print(intervals.head())


if __name__ == "__main__":
    main()
```
